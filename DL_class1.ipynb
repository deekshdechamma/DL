# Movie Review Sentiment Analysis using Simple RNN and Bidirectional RNN


```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, Bidirectional
from sklearn.metrics import classification_report
```

```python
# Load IMDb Dataset

vocab_size = 10000
max_length = 200

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=vocab_size)
```

```python
# Pad sequences
X_train = pad_sequences(X_train, maxlen=max_length)
X_test = pad_sequences(X_test, maxlen=max_length)
```

```python
# Simple RNN Model

print("\nTraining Simple RNN Model...\n")

simple_rnn = Sequential([
    Embedding(input_dim=vocab_size, output_dim=128, input_length=max_length),
    SimpleRNN(64),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

simple_rnn.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

simple_rnn.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)

loss_rnn, acc_rnn = simple_rnn.evaluate(X_test, y_test, verbose=0)

print("\nSimple RNN Results")
print("------------------")
print(f"Test Loss     : {loss_rnn:.4f}")
print(f"Test Accuracy : {acc_rnn*100:.2f}%")
```

```python
# Predictions
pred_rnn = (simple_rnn.predict(X_test) > 0.5).astype(int)

print("\nClassification Report (Simple RNN)")
print(classification_report(y_test, pred_rnn))

```

```python
# Bidirectional RNN Model

print("\nTraining Bidirectional RNN Model...\n")

bi_rnn = Sequential([
    Embedding(input_dim=vocab_size, output_dim=128, input_length=max_length),
    Bidirectional(SimpleRNN(64)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

bi_rnn.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

bi_rnn.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)

loss_bi, acc_bi = bi_rnn.evaluate(X_test, y_test, verbose=0)

print("\nBidirectional RNN Results")
print("-------------------------")
print(f"Test Loss     : {loss_bi:.4f}")
print(f"Test Accuracy : {acc_bi*100:.2f}%")

# Predictions
pred_bi = (bi_rnn.predict(X_test) > 0.5).astype(int)

print("\nClassification Report (Bidirectional RNN)")
print(classification_report(y_test, pred_bi))
```

```python
# Model Comparison

print("\n===================================")
print("MODEL COMPARISON")
print("===================================")

print(f"Simple RNN Accuracy       : {acc_rnn*100:.2f}%")
print(f"Bidirectional RNN Accuracy: {acc_bi*100:.2f}%")

if acc_bi > acc_rnn:
    print("\nConclusion:")
    print("Bidirectional RNN performs better than Simple RNN for IMDb sentiment analysis.")
else:
    print("\nConclusion:")
    print("Simple RNN performs better than Bidirectional RNN.")
```
